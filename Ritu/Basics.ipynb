Grey Scaling

```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('ecosort.jpg')
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python
grey_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(grey_img, cmap='gray')
plt.axis('off')
plt.show()
```

Thresholding

```python
plt.imshow(grey_img, 'gray', vmin=0, vmax=255)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
ret, thresh_binary = cv2.threshold(grey_img, 127, 255, cv2.THRESH_BINARY)

plt.imshow(thresh_binary, 'gray', vmin=0, vmax=255)
plt.title('BINARY')
plt.axis('off')
plt.show()
```

```python
ret, thresh_binary_inv = cv2.threshold(grey_img, 127, 255, cv2.THRESH_BINARY_INV)

plt.imshow(thresh_binary_inv, 'gray', vmin=0, vmax=255)
plt.title('BINARY_INV')
plt.axis('off')
plt.show()
```

```python
ret, thresh_trunc = cv2.threshold(grey_img, 127, 255, cv2.THRESH_TRUNC)

plt.imshow(thresh_trunc, 'gray', vmin=0, vmax=255)
plt.title('TRUNC')
plt.axis('off')
plt.show()
```

```python
ret, thresh_tozero_inv = cv2.threshold(grey_img, 127, 255, cv2.THRESH_TOZERO_INV)

plt.imshow(thresh_tozero_inv, 'gray', vmin=0, vmax=255)
plt.title('TOZERO_INV')
plt.axis('off')
plt.show()
```

```python
edges = cv2.Canny(grey_img, 100, 200)
```

```python
plt.imshow(edges,cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')
plt.show()
```
