```python
img =cv2.imread('WhatsApp Image 2024-09-19 at 6.56.20 PM.jpeg')
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
gray_img =cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img,cmap='gray')
plt.axis("off")
plt.show()
ret,thresh_binary=cv2.threshold(gray_img,127,255,cv2.THRESH_TRUNC)
plt.imshow(thresh_binary,'gray',vmin=0,vmax=100)
plt.axis("off")
plt.show()
ret,thresh_binary_inv=cv2.threshold(gray_img,127,255,cv2.THRESH_BINARY_INV)
plt.imshow(thresh_binary_inv,'gray',vmin=0,vmax=100)
plt.axis("off")
plt.show()
ret,thresh_trunc=cv2.threshold(gray_img,127,255,cv2.THRESH_BINARY)
plt.imshow(thresh_trunc,'gray',vmin=0,vmax=100)
plt.axis("off")
plt.show()
ret,thresh_tozero=cv2.threshold(gray_img,127,255,cv2.THRESH_TOZERO)
plt.imshow(thresh_tozero,'gray',vmin=0,vmax=100)
plt.axis("off")
plt.show()
ret,thresh_tozero_inv=cv2.threshold(gray_img,127,255,cv2.THRESH_TOZERO_INV)
plt.imshow(thresh_tozero_inv,'gray',vmin=0,vmax=100)
plt.axis("off")
plt.show()
```

```python
import numpy as np
from sklearn.cluster import KMeans
```

```python
pixel_values=img.reshape((-1,3))
pixel_values=np.float32(pixel_values)
```

```python

k=4
kmeans=KMeans(n_clusters=k,random_state=42)
kmeans.fit(pixel_values)
labels=kmeans.labels_
centers=kmeans.cluster_centers_
centers_centers=np.uint8(centers)
```

```python
segmented_image= centers[labels.flatten()]
segmentented_image=segmented_image.reshape(img.shape)
plt.imshow(segmented_image)
plt.title(f'segmented Image with {k} means')
plt.show()
```

```python
for k in [2,4,6]:
 kmeans=KMeans(n_clusters=k,random_state=42)
 kmeans.fit(pixel_values)
 centers=np.uint8(kmeans.cluster_centers_)
segmented_image=centers[kmeans.labels_.flatten()].reshape(img.shape)

plt.imshow(segmented_image)
plt.title(f'segmented Image with {k} colors')
plt.show()

    
```
